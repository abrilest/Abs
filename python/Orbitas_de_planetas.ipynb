```python
import os
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
def solar(x0, y0, z0, vx0, vy0, vz0, t0=0, dt=0.01 ,nsteps=100):
    G = 4*(np.pi**2)
    M = 1
    rp = np.sqrt( x0*2 + y0**2 + z0**2 ) 
    x = [x0] 
    y = [y0]
    z = [z0]
    t = [t0]
    for i in range(1, nsteps):
        Vxi = vx0 - ( (G*M) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        Vyi = vy0 - ( (G*M) / rp**3 )*y0 *dt 
        yi = y0 + Vyi*dt
        Vzi = vz0 - ( (G*M) / rp**3 )*z0 *dt 
        zi = z0 + Vzi*dt
        x.append(xi)
        y.append(yi)
        z.append(zi)
        t.append(t0+dt*i)
        x0 = xi
        vx0 = Vxi 
        y0= yi
        vy0 = Vyi
        z0 = zi
        vz0 = Vzi
        
    return x, y, z, t
```

```python
x,y,z,t=solar(0.0508011, 0.022318, -0.4011642, -0.2195576,0.00485553,0.00027924,nsteps=100)
plt.plot(t,x)
```
